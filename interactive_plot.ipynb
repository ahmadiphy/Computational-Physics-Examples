```python
%matplotlib notebook
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x=5.,y=5.):
    return x**3+y**2
```

```python
interact(f);
```

```python
x = np.linspace(0, 2*np.pi)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, np.sin(x),linestyle=":",marker='o',markersize=5,markerfacecolor='red')
def update(a = 1.0,b=1.0):
    line.set_ydata(np.sin(a * x+b))
    fig.canvas.draw()
interact(update);
```
